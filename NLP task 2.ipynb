```python
import nltk
from nltk.tokenize import word_tokenize
from nltk import pos_tag, ne_chunk
from nltk.classify import NaiveBayesClassifier
from nltk.metrics import precision
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams
from nltk.sem import Expression

nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('averaged_perceptron_tagger_eng')
nltk.download('maxent_ne_chunker')
nltk.download('maxent_ne_chunker_tab')
nltk.download('words')

news = """
Google launched a new cloud platform in Bengaluru.
The company believes the technology will improve business performance and innovation.
"""

tokens = word_tokenize(news)
pos_tags = pos_tag(tokens)

for word, tag in pos_tags:
    if word == "Google":
        print("Subject :", word)
    elif word == "launched":
        print("Verb :", word)

print("Object : cloud platform")

print("\n=== Named Entities ===")

chunked = ne_chunk(pos_tags)

for subtree in chunked:
    if hasattr(subtree, 'label'):
        entity = " ".join(c[0] for c in subtree)
        print(subtree.label(), ":", entity)

def features(text):
    words = word_tokenize(text.lower())
    return {word: True for word in words}

train_data = [
    (features("football cricket tournament"), "Sports"),
    (features("election parliament government"), "Politics"),
    (features("market finance business"), "Business"),
    (features("software computer technology ai"), "Technology")
]

classifier = NaiveBayesClassifier.train(train_data)

category = classifier.classify(features(news))

print("\nCategory :", category)

test_set = [
    (features("software ai technology"), "Technology"),
    (features("football cricket"), "Sports")
]

acc = nltk.classify.accuracy(classifier, test_set)

print("\n=== Model Accuracy ===")
print(round(acc * 100, 2), "%")

refset = {'Technology'}
testset = {category}

print("Precision :", precision(refset, testset))

words = [w.lower() for w in word_tokenize(news) if w.isalpha()]

fdist = FreqDist(words)

print("\n=== Word Frequency ===")

for word, freq in fdist.items():
    print(word, ":", freq)

print("\n=== Probability Distribution ===")

total_words = len(words)

for word in fdist:
    print(f"P({word}) = {fdist[word]}/{total_words} = {fdist[word]/total_words:.2f}")

read_expr = Expression.fromstring

expr1 = read_expr('technology(innovation)')
expr2 = read_expr('cloud(technology)')

print("\n=== Semantic Analysis ===")
print("Semantic Expression 1 :", expr1)
print("Semantic Expression 2 :", expr2)

print("\n=== Bigrams ===")

for bg in bigrams(words):
    print(bg)

print("\n=== Trigrams ===")

for tg in trigrams(words):
    print(tg)
```
